```python
from huggingface_hub import login
from diffusers import StableDiffusionPipeline
from google.colab import files
login(HF_TOKEN)  # Hugging Face'e giriş yap
```

```python
# Eğitilen modeli yükle
pipe = StableDiffusionPipeline.from_pretrained("firatyapayzeka/turkoglu")
pipe.to("cuda")  # GPU kullanımı
```

```python
# Prompt ile görsel üretimi
prompt ="A realistic photo of turkoglu driving a modern car, with a cityscape visible through the windows, and a focused expression on his face."















negative_prompt = ""


generated_image = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]

# Görüntüyü kaydet
generated_image.save("generated_image.png")
generated_image.show()

files.download("generated_image.png")
```
